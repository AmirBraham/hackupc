```python
import requests
import re
import os
from bs4 import BeautifulSoup
```

```python
def file_exists(file_path):
    return os.path.exists(file_path) and os.path.getsize(file_path) > 0
def clean_text(text):
    # Replace non-alphabetical characters with space
    cleaned_text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    return cleaned_text
```

```python
def get_main_text(content):
    # Parse the HTML content of the webpage
    soup = BeautifulSoup(content, "html.parser")
    
    # Find the element with class "main"
    main_element = soup.find(class_="main")
    
    # Extract text from the main element
    if main_element:
        main_text = main_element.get_text()
        return main_text
    else:
        print("No element with class 'main' found.")
        return ""
```

```python
def get_title(content):
    soup = BeautifulSoup(content, "html.parser")
    
    # Find the element with class "main"
    main_element = soup.find(class_="with-also")
    
    # Extract text from the main element
    if main_element:
        main_text = main_element.get_text()
        return main_text
    else:
        print("No element with class 'with_also' found.")
        return ""
```
